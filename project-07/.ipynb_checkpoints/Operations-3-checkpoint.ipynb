#### Hub Type: Percentage of Annual Passenger Boardings

* Large Hub: “large hub airport” means a commercial service airport that has at least 1.0 percent of the passenger boardings.
* Medium Hub: “medium hub airport” means a commercial service airport that has at least 0.25 percent but less than 1.0 percent of the passenger boardings.
* Small Hub: “small hub airport” means a commercial service airport that has at least 0.05 percent but less than 0.25 percent of the passenger boardings.
* Nonhub Primary: “nonhub airport” means a commercial service airport that has less than 0.05 percent of the passenger boardings.

https://www.faa.gov/airports/planning_capacity/passenger_allcargo_stats/categories/

```python
print Airport_passengers_04_14_df.head(1)
```

```python
passenger_col = []
airport_col = []
year = []
years = ['2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014']
for i in range(1558):
    row = list(airport_passengers_04_14_temp_df.ix[i][1:])
    passenger_col = np.concatenate((passenger_col, row))
    airport = [airport_passengers_04_14_temp_df.ix[i][0]]*11
    airport_col = np.concatenate((airport_col, airport))
    year = np.concatenate((year, years))
passenger_data = [airport_col, year, passenger_col]
airport_passengers_04_14_df = pd.DataFrame(passenger_data, index=('Airport', 'year', 'passengers')).T
print airport_passengers_04_14_df.shape
airport_passengers_04_14_df.head()
```

## Merging

```python
passenger_cargo_year = pd.merge(airport_passengers_04_14_df, airport_cargo_04_14_df, how='left', on=['Airport', 'year'])
print passenger_cargo_year.shape
passenger_cargo_year.year = passenger_cargo_year.year.astype(int)
passenger_cargo_year
```

```python
print airport_operations_04_14_df.shape
airport_operations_04_14_df.head()
```

```python
operations_passenger_cargo_year = airport_operations_04_14_df.merge(passenger_cargo_year, how='left',\
                                           left_on=['Airport', 'year'], right_on=['Airport', 'year'])
operations_passenger_cargo_year
```

```python
Airport_operations_df.head()
```

```python
airport_operations_df = Airport_operations_df.merge(operations_passenger_cargo_year, how='left',\
                                           left_on=['airport', 'year'], right_on=['Airport', 'year'])
airport_operations_df
```
