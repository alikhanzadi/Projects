##### 13. Write a function using only list comprehensions, no loops, to compute Standard Deviation. Print the Standard Deviation of each numeric column.

```python
import numpy as np
def stdv (numeric_column):
    average = np.mean(numeric_column.values())
    sum_dif_sq = sum([(cell - average)**2 for cell in numeric_column.values()])/len(numeric_column.values())
    std = sum_dif_sq**.5
    return std
print ('Rate Standard Deviation is ', stdv (State_Rate))
print ('Verbal Standard Deviation is ', stdv (State_Verbal))
print ('Math Standard Deviation is ', stdv (State_Math))
```
